**Problem - 1: Getting Started with Data Exploration**

```python
# 1. Data Exploration and Understanding:

# Dataset Overview:
# 1.Load the dataset and display the first 10 rows.
import pandas as pd
dataset = pd.read_csv('/content/drive/MyDrive/Assigment1/WHR-2024-5CS037.csv')
print("Fist 10 rows: \n",dataset.head(10))

# 2.Identify the number of rows and columns in the dataset.
rows, columns = dataset.shape
print(f"\nNumber of rows: {rows}")
print(f"Number of columns: {columns}")

# 3.List all the columns and their data types.
print("\nColumns and their data types: \n",dataset.dtypes)
```

```python
# Basic Statistics:

# 1.Calculate the mean, median, and standard deviation for the Score column.
print("Mean: ",dataset['score'].mean())
print("Median: ",dataset['score'].median())
print("Standard Deviation: ",dataset['score'].std())

# 2.Identify the country with the highest and lowest happiness scores.
print("\nCountry with the highest Happiness Score: ",dataset.loc[dataset['score'].idxmax(), 'Country name'])
print("Country with the lowest Happiness Score: ",dataset.loc[dataset['score'].idxmin(), 'Country name'])
```

```python
# Missing Values:

# 1.Check if there are any missing values in the dataset.
print("Missing Values:\n",dataset.isnull().sum())
```

```python
# Filtering and Sorting

# 1.Filter the dataset to show only the countries with a Score greater than 7.5.
filtered_dataset = dataset[dataset['score'] > 7.5]
print("Filtered Dataset:\n", filtered_dataset)

# 2.For the filtered dataset - Sort the dataset by 'Log GDP per capita' in descending order and display the top 10 rows.
sorted_dataset = filtered_dataset.sort_values(by='Log GDP per capita', ascending=False)
print("\nTop 10 Countries by Log GDP per Capita:\n", sorted_dataset.head(10))
```

```python
# Adding New Columns:
import numpy as np
# 1.Create a new column Happiness Category based on the Score column.
conditions = [
    (dataset['score'] < 4),
    (dataset['score'] >= 4) & (dataset['score'] <= 6),
    (dataset['score'] > 6)
]
categories = ['Low', 'Medium', 'High']
dataset['Happiness Category'] = np.select(conditions, categories)
print(dataset[['Country name', 'Happiness Category']])
```

```python
# 2. Data Visualizations:

# Bar Plot
import matplotlib.pyplot as plt

happiest_countries = dataset.sort_values(by='score', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.barh(happiest_countries['Country name'], happiest_countries['score'], color='lightblue')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.title('Top 10 Happiest Countries by Score')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Line Plot:
import matplotlib.pyplot as plt

unhappiest_countries = dataset.sort_values(by='score', ascending=True).head(10)
plt.figure(figsize=(10, 6))
plt.plot(unhappiest_countries['Country name'], unhappiest_countries['score'], marker='o', color='lightblue')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.title('Top 10 Unhappiest Countries by Score')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

```python
# Task - 3 - Outlier Detection:

# 1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
Q1_score = south_asian_dataset['score'].quantile(0.25)
Q3_score = south_asian_dataset['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asian_dataset['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asian_dataset['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp


# 2. Define outliers using the 1.5  IQR rule
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

outliers_score = south_asian_dataset[(south_asian_dataset['score'] < lower_bound_score) | (south_asian_dataset['score'] > upper_bound_score)]
outliers_gdp = south_asian_dataset[(south_asian_dataset['Log GDP per capita'] < lower_bound_gdp) | (south_asian_dataset['Log GDP per capita'] > upper_bound_gdp)]

outliers_combined = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()
print("Outliers based on Score and GDP per Capita:\n", outliers_combined[['Country name', 'score', 'Log GDP per capita']])
```

```python
# 3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
plt.figure(figsize=(10, 6))

plt.scatter(south_asian_dataset['Log GDP per capita'], south_asian_dataset['score'], color='blue', label='Normal')

plt.scatter(outliers_combined['Log GDP per capita'], outliers_combined['score'], color='red', label='Outliers', zorder=5)

plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('GDP per Capita vs. Score with Outliers Highlighted', fontsize=14)
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
